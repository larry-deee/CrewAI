Example with no Logging

```python
# CrewAI Imports
from crewai import Agent, Task, Crew, LLM, Process
from crewai_tools import PDFSearchTool, FileReadTool
from google.colab import userdata
import os
import json
import re

# Set up environment variables
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
os.environ["ANTHROPIC_API_KEY"] = userdata.get('ANTHROPIC_API_KEY')
os.environ["HF_TOKEN"] = userdata.get('HF_TOKEN')

# Initialize LLM and tools
#llm = LLM(model="anthropic/claude-3-5-haiku-latest", temperature=0.55)
llm = LLM(model="anthropic/claude-3-5-sonnet-latest", temperature=0.55, max_tokens=4000)
#llm = LLM(model="openai/gpt-4o-2024-11-20", temperature=0.55)

file_read_tool = FileReadTool(file_path="/content/sample.txt")

# Ingest PDFs
pdf_paths = [
    "/content/2201_08675.pdf",
    "/content/guidelines_gender-sensitive_language_e-a.pdf",
    "/content/gender_shades.pdf"
]

pdf_tools = []
for path in pdf_paths:
    try:
        pdf_tool = PDFSearchTool(pdf=path)
        pdf_tools.append(pdf_tool)
        print(f"Successfully ingested PDF: {path}")
    except Exception as e:
        print(f"Error ingesting PDF {path}: {e}")

# Define Agents
document_reader = Agent(
    name="DocumentReader",
    role="Document Reading Agent",
    goal="Read and process input text files accurately",
    backstory="An AI expert in processing and extracting text from files",
    tools=[file_read_tool],
    llm=llm
)

lexicon_extractor = Agent(
    name="LexiconExtractor",
    role="Lexicon Extraction Agent",
    goal="Extract structured lexicons for gender bias and stereotypes from PDFs.",
    backstory="An AI expert skilled in processing and structuring large volumes of text data.",
    tools=pdf_tools + [file_read_tool],
    llm=llm
)

document_comparer = Agent(
    name="DocumentComparer",
    role="Document Comparison Agent",
    goal="Identify and flag problematic terms based on lexicons.",
    backstory="A meticulous copy editor with expertise in identifying bias.",
    llm=llm
)

text_rewriter = Agent(
    name="TextRewriter",
    role="Rewriting Agent",
    goal="Rewrite flagged terms with inclusive alternatives while maintaining context.",
    backstory="A skilled writer ensuring clear, inclusive communication.",
    llm=llm
)

text_validator = Agent(
    name="TextValidator",
    role="Validation Agent",
    goal="Review rewritten text for approval.",
    backstory="An expert ensuring compliance and accuracy in written materials.",
    tools=[file_read_tool],
    llm=llm
)

# Define Tasks
read_task = Task(
    description=f"Read the input text file from {file_read_tool}. Confirm you have read the file.",
    expected_output="Raw text content from the file",
    verbose=True,
    agent=document_reader
)

extract_task = Task(
    description="Extract lexicons for gender bias and stereotypes from the PDFs.",
    expected_output="JSON formatted lexicons.",
    output_file="lexicons.json",
    agent=lexicon_extractor,
    context=[read_task]
)

compare_task = Task(
    description="Compare the input document text against the lexicons.",
    expected_output="Flagged terms with suggestions for inclusive language.",
    output_file="flagged_terms.json",
    verbose=True,
    agent=document_comparer,
    context=[read_task, extract_task]
)

rewrite_task = Task(
    name='rewrite_task',
    description="Rewrite the original input document text into a compliant version.",
    expected_output="""
        A complete rewritten version of the original text with all applicable modifications
        applied from the lexicon. Ensure your output is a variation of the original text
    """,
    output_file="sample_fixed.txt",
    agent=text_rewriter,
    context=[read_task, compare_task]
)

validate_task = Task(
    description="Validate the original vs. the newly rewritten document.",
    expected_output="Ensure the rewritten text is in compliance.",
    tools=[file_read_tool],
    agent=text_validator,
    context=[read_task, rewrite_task]
)

# Define and Execute Crew
bias_detection_crew = Crew(
    agents=[
        document_reader,
        lexicon_extractor,
        document_comparer,
        text_rewriter,
        text_validator
    ],
    tasks=[
        read_task,
        extract_task,
        compare_task,
        rewrite_task,
        validate_task
    ],
    memory=True,
    process=Process.sequential,
    #max_rpm=5,
    verbose=True
)

# Execute the crew
results = bias_detection_crew.kickoff()
print("Final Output:", results)
```

Example with Logging

```python
# CrewAI Imports
from crewai import Agent, Task, Crew, LLM, Process
from crewai_tools import PDFSearchTool, FileReadTool
from google.colab import userdata
from datetime import datetime
from pathlib import Path
import os
import json
import logging

# Initialize directories and logging
log_dir = Path('/content/logs')
log_dir.mkdir(exist_ok=True)

logger = logging.getLogger('bias_detection_logger')
logger.setLevel(logging.DEBUG)

formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handlers = [
    logging.FileHandler(log_dir / f'bias_detection_audit_{datetime.now().strftime("%Y%m%d_%H%M%S")}.log'),
    logging.StreamHandler()
]
for handler in handlers:
    handler.setFormatter(formatter)
    logger.addHandler(handler)

# Environment setup
for key in ['OPENAI_API_KEY', 'ANTHROPIC_API_KEY', 'HF_TOKEN']:
    os.environ[key] = userdata.get(key)

# Initialize LLM and tools
llm = LLM(model="openai/gpt-4o-2024-11-20", temperature=0.55)
file_read_tool = FileReadTool(file_path="/content/sample.txt")

# Ingest PDFs
pdf_paths = [
    "/content/2201_08675.pdf",
    "/content/guidelines_gender-sensitive_language_e-a.pdf",
    "/content/gender_shades.pdf"
]
pdf_tools = [PDFSearchTool(pdf=path) for path in pdf_paths]

def step_callback(step_output):
    logger.info(f"Step Output: {str(step_output)}")

def usage_metrics_to_dict(metrics):
    return {
        attr: getattr(metrics, attr, 0)
        for attr in ['prompt_tokens', 'completion_tokens', 'total_tokens', 'cost']
    } if metrics else {}

class CrewMetricsEncoder(json.JSONEncoder):
    def default(self, obj):
        if hasattr(obj, '__dict__'):
            return {k: v for k, v in obj.__dict__.items()
                   if not k.startswith('_') and not callable(v)}
        return str(obj)

def create_tableau_metrics(crew, start_time):
    return {
        "execution_metrics": {
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "run_id": start_time.strftime("%Y%m%d_%H%M%S"),
            "duration": (datetime.now() - start_time).total_seconds(),
            "process_type": str(crew.process),
            "agents_count": len(crew.agents),
            "tasks_count": len(crew.tasks),
            "usage_metrics": usage_metrics_to_dict(crew.usage_metrics),
            "tasks": [{
                "name": task.description[:50] + "...",
                "agent": task.agent.role,
                "status": "completed" if hasattr(task, 'output') else "pending",
                "output_file": getattr(task, 'output_file', None)
            } for task in crew.tasks]
        },
        "agent_metrics": [{
            "role": agent.role,
            "goal": agent.goal,
            "tools_count": len(agent.tools or []),
            "tools": [getattr(tool, 'name', str(tool)) for tool in (agent.tools or [])]
        } for agent in crew.agents],
        "validation_status": "passed" if hasattr(crew.tasks[-1], 'output') and
            "compliant" in str(crew.tasks[-1].output.raw).lower() else "needs_review"
    }

# Define Agents
agents = {
    "document_reader": Agent(
        name="DocumentReader",
        role="Document Reading Agent",
        goal="Read and process input text files accurately",
        backstory="An AI expert in processing and extracting text from files",
        tools=[file_read_tool],
        llm=llm
    ),
    "lexicon_extractor": Agent(
        name="LexiconExtractor",
        role="Lexicon Extraction Agent",
        goal="Extract structured lexicons for gender bias and stereotypes from PDFs.",
        backstory="An AI expert skilled in processing and structuring large volumes of text data.",
        tools=pdf_tools + [file_read_tool],
        llm=llm
    ),
    "document_comparer": Agent(
        name="DocumentComparer",
        role="Document Comparison Agent",
        goal="Identify and flag problematic terms based on lexicons.",
        backstory="A meticulous copy editor with expertise in identifying bias.",
        llm=llm
    ),
    "text_rewriter": Agent(
        name="TextRewriter",
        role="Rewriting Agent",
        goal="Rewrite flagged terms with inclusive alternatives while maintaining context.",
        backstory="A skilled writer ensuring clear, inclusive communication.",
        llm=llm
    ),
    "text_validator": Agent(
        name="TextValidator",
        role="Validation Agent",
        goal="Review rewritten text for approval.",
        backstory="An expert ensuring compliance and accuracy in written materials.",
        tools=[file_read_tool],
        llm=llm
    )
}

# Define Tasks
# Store tasks in a list as we create them for referencing
tasks = []

# First task (no context needed)
tasks.append(Task(
    description=f"Read the input text file from {file_read_tool}. Confirm you have read the file.",
    expected_output="Raw text content from the file",
    verbose=True,
    output_file="logs/read_task_output.txt",
    agent=agents["document_reader"]
))

# Add subsequent tasks with proper context references
tasks.append(Task(
    description=f"Extract highly detailed, audit compliant lexicons for gender bias and stereotypes from the reference documents.",
    expected_output=f"JSON formatted lexicon derived from {pdf_tools}.",
    output_file="logs/lexicons.json",
    verbose=True,
    agent=agents["lexicon_extractor"],
    context=[tasks[0]]
))

tasks.append(Task(
    description="Compare the input document text against the lexicon data.",
    expected_output="Flagged terms with suggestions for inclusive language based on information from lexicon data.",
    output_file="logs/flagged_terms.json",
    verbose=True,
    agent=agents["document_comparer"],
    context=[tasks[0], tasks[1]]
))

tasks.append(Task(
    description="Rewrite the original input document text into a compliant version.",
    expected_output="A complete rewritten version of the original text with all applicable modifications applied from the lexicon.",
    output_file="logs/sample_fixed.txt",
    verbose=True,
    agent=agents["text_rewriter"],
    context=[tasks[0], tasks[2]]
))

tasks.append(Task(
    description="Validate the original vs. the newly rewritten document.",
    expected_output="Ensure the rewritten text is in compliance.",
    verbose=True,
    agent=agents["text_validator"],
    context=[tasks[0], tasks[3]]
))

# Execute Crew
bias_detection_crew = Crew(
    agents=list(agents.values()),
    tasks=tasks,
    memory=True,
    process=Process.sequential,
    verbose=True,
    step_callback=step_callback
)

start_time = datetime.now()
try:
    logger.info("Starting bias detection process")
    results = bias_detection_crew.kickoff()

    # Generate metrics
    tableau_metrics = create_tableau_metrics(bias_detection_crew, start_time)

    # Save metrics to JSON
    metrics_file = log_dir / f'tableau_metrics_{start_time.strftime("%Y%m%d_%H%M%S")}.json'
    metrics_file.write_text(json.dumps(tableau_metrics, indent=2, cls=CrewMetricsEncoder))

    logger.info(f"Results: {results.raw}")
    logger.info(f"Metrics saved to: {metrics_file}")

except Exception as e:
    logger.error(f"Error during execution: {str(e)}")
```
